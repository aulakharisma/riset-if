```python
movie_pivot=df.pivot_table(columns='userId', index='title', values='rating')
```

```python
movie_pivot.shape
```

```python
movie_pivot.fillna(0,inplace=True)
```

```python
movie_pivot
```

```python
from scipy.sparse import csr_matrix
movie_sparse = csr_matrix(movie_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(n_neighbors=7, algorithm='brute', metric='cosine')
```

```python
model.fit(movie_sparse)
```

```python
df.drop(columns=['genres','userId','rating'],inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
def reco(movie_name):
    try:
        movie_data = df1[df1['title'] == movie_name].drop_duplicates('title')

        if movie_data.empty:
            return "Film tidak ditemukan."

        movie_id = movie_data['movieId'].values[0]
        distances, suggestions = model.kneighbors(movie_pivot.iloc[movie_id, :].values.reshape(1, -1))

        recommended_indices = []
        for i in range(min(5, len(suggestions[0]))):
            if suggestions[0][i] < len(movie_pivot.index):
                recommended_indices.append(movie_pivot.index[suggestions[0][i]])

        return recommended_indices
    except IndexError:
        return "Tidak ada rekomendasi yang ditemukan."
```

```python
recommended_movies = reco('Avatar')
print(recommended_movies)
```
